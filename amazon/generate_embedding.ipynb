```python
from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import euclidean_distances
import pandas as pd
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. Load the lightweight model
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Assume 'product_metadata' is your list of 100k strings
# Process in batches to keep the UI/system responsive
embeddings = model.encode([
        "Mamaearth Hydragel Indian Sunscreen Spf 50, With Aloe Vera & Raspberry, For Sun Protection - 50G",
        "Mamaearth Gentle Sunscreen for Babies (200 ml)",
        'Giraffe Genie Glitter Backpack for Women - 3 compartments, Laptop Bags for Girls, Stylish and Trendy College Backpacks for Girls, ...',
        'Coolz Kids Chu-Chu Sound Musical First Walking Sandals C-06 for Baby Boys and Baby Girls Age 12-24 Months',
    ], 
    batch_size=64, 
    show_progress_bar=True, 
    convert_to_numpy=True
)

# 3. L2-Normalize for your metric requirement
embeddings = embeddings / np.linalg.norm(embeddings, axis=1, keepdims=True)

# 4. Save to disk so you never have to do it again!
np.save("product_vectors.npy", embeddings)
```

```python
# 4. Compute Pairwise Metric (Euclidean Distance)
dist_matrix = euclidean_distances(embeddings) / np.sqrt(2)

# 5. Display as a clean DataFrame
df_pairwise = pd.DataFrame(dist_matrix)

# Format to 3 decimal places for readability
print(df_pairwise.round(3).to_string())
```
